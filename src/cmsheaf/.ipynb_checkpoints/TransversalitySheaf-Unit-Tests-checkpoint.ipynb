```python
def cusp ( x, params ):
    a,b = params
    return -x**3 + a*x + b
#Bifurcations (Pitchfork, Saddle-node)
def PF ( x, a ):
    return -x**3 + a*x
def SN (x, a ):
    return x**2 + a
def hopf ( X, a):
    sigma = -1.
    #a,sigma = params
    return [a*X[0]-X[1]+sigma*X[0]*(X[0]**2+X[1]**2), 
            X[0]+a*X[1]+sigma*X[1]*(X[0]**2+X[1]**2)]
```
